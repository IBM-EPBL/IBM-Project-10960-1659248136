MODEL BUILDING

ADDING LIBRARY FILES

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
```

IMPORTING DATASET

```python
dat=pd.read_csv('./car_performance.csv')
dat
```

CHECKING FOR MISSING DATA

```python
dat.isnull().any()
```

VISUALIZATION

PAIRPLOT

```python
sns.pairplot(dat,diag_kind='kde')
plt.show()
```

HEATMAP

```python
sns.heatmap(dat.corr(),annot=True,linecolor ='black', linewidths = 1)
fig=plt.gcf()
fig.set_size_inches(5,5)
```

SEPERATING THE DATA

```python
x=dat[['cylinders','displacement','horsepower','weight','model year','origin']].values #INDEPENDENT VARIABLE
x
```

```python
y=dat.iloc[:,0:1].values #DEPENDENT VARIABLE
y
```

SPLITTING TEAT AND TRAIN DATA

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=0)
```

TESTING THE ALGORITHMS

1. LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred1=reg.predict(x_test)
y_pred1
```

```python
ax1 = sns.distplot(dat['mpg'], hist=False, color="r", label="Actual Value")
sns.distplot(y_pred1, hist=False, color="b", label="Fitted Values" , ax=ax1)

plt.title('Actual vs Fitted Values for mpg')
plt.xlabel('mpg')
plt.ylabel('Proportion of Cars')

plt.show()
plt.close()
```

CALCULATING METRICS

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y_test,y_pred1)
```

```python
mean_squared_error(y_test,y_pred1)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred1))
```

2. USING RANDOM FOREST REGRESSOR

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf= RandomForestRegressor(n_estimators=10,random_state=0,criterion='mae')
rf.fit(x_train,y_train)
```

```python
y_pred2=rf.predict(x_test)
y_pred2
```

```python
ax1 = sns.distplot(dat['mpg'], hist=False, color="r", label="Actual Value")
sns.distplot(y_pred2, hist=False, color="b", label="Fitted Values" , ax=ax1)

plt.title('Actual vs Fitted Values for mpg')
plt.xlabel('mpg')
plt.ylabel('Proportion of Cars')

plt.show()
plt.close()
```

CALCULATING METRICS

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y_test,y_pred2)
```

```python
mean_squared_error(y_test,y_pred2)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred2))
```

3. DECESION TREE REGRESSION

```python
from sklearn.tree import DecisionTreeRegressor   
```

```python
dt = DecisionTreeRegressor(random_state = 0) 
dt.fit(x_train, y_train)
```

```python
y_pred3=dt.predict(x_test)
y_pred3
```

```python
ax1 = sns.distplot(dat['mpg'], hist=False, color="r", label="Actual Value")
sns.distplot(y_pred3, hist=False, color="b", label="Fitted Values" , ax=ax1)

plt.title('Actual vs Fitted Values for mpg')
plt.xlabel('mpg')
plt.ylabel('Proportion of Cars')

plt.show()
plt.close()
```

CALCULATING METRICS

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y_test,y_pred3)
```

```python
mean_squared_error(y_test,y_pred3)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred3))
```

CONCLUSION

The Algorithm with higher R score and lower MSE is the best algorithm.
